```python
# Sachin tendulkar's runs over the years
# Adding details to my visualization
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
plt.show()
```

## Adding X and Y labels

```python
plt.plot(years,runs)
plt.xlabel("Year")
plt.ylabel("RUns Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

## Multiple Lines in One Plot

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kohli,label = "Virat Kohli")
plt.plot(years,sehwag,label = "sehwag")

plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance comparison")
plt.legend()

# It's a bit difficult to visualizae the insights just looking at the visualizations
```

## Why label is Better than List in legend

```python
# Because it can create confusions if i have a lot of data. What if i have 10 different data, so creating a legend 
# using a list is messy. I need to look at each data and then write the corresponding legend for that data.
```

## Using Format Strings


```python
plt.plot(years, kohli, 'r|--',label = "Kohli") #Red Dashed lines
plt.plot(years,sehwag,'g^:',label = "Sehwag") #Green triangles
# I can use many combinations of color marker and then line[color][marker][line]
```

```python
## Color and Line Style Arguments
plt.plot(years,kohli,color='orange',linestyle = "--",label = "kohli",linewidth = 2)
plt.plot(years,sehwag,color='green',linestyle="-.",label = "sehwag",linewidth = 2)
plt.plot(years,runs,color = 'red',linestyle=":",label = "tendulkar",linewidth = 2)
plt.legend()
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance comparison")
plt.grid(True)
plt.tight_layout()
```

## Line Width and Layout Tweaks

```python
plt.plot(years, kohli, linewidth=3, label="Kohli")
plt.plot(years, sehwag, linewidth=2, label="Sehwag")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## XKCD Comic Style

```python
import matplotlib.pyplot as plt
with plt.xkcd():
    plt.rcParams["font.family"] = "DejaVu Sans"
    plt.plot(years,kohli,label="Kohli")
    plt.plot(years,sehwag,label="sehwag")
    plt.title("Battle between the legends")
    plt.legend()
```
